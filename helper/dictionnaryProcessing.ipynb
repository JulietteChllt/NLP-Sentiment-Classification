## Harvard dictionnary

```python
"""
Load xls
"""
PATH_TRAIN = '../Resources/initialDictionnaries/inquirerbasic.xls'
```

```python
"""
read xls
"""
data_harvard = pd.read_excel(PATH_TRAIN)
```

```python
data_harvard
```

```python
"""
prepare harvard dictionnary: remove useless columns
"""
data_harv=data_harvard[["Entry","Positiv","Negativ"]]
```

```python
"""
create two lists, one with the positive words and one with the negative ones
"""
data_harv_pos = data_harv.loc[data_harv['Positiv'].isnull()==False]
data_harv_neg = data_harv.loc[data_harv['Negativ'].isnull()==False]
```

```python
"""
remove all column but the one containing the word 
"""
data_harv_positive = data_harv_pos[["Entry"]]
data_harv_negative = data_harv_neg[["Entry"]]
```

```python
"""
save dictionnaries as csv
"""
data_harv_positive.to_csv('C:/Users/julie/OneDrive/Documents/GitHub/MLProject2/Resources/harvard_pos.csv', header = False, index=False)
data_harv_negative.to_csv('C:/Users/julie/OneDrive/Documents/GitHub/MLProject2/Resources/harvard_neg.csv', header = False, index=False)
```
